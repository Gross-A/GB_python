ниже функция для поиска корней: нужно задать интервал поиска и указать убывает или возрастает функция на этом участке (исходя из графика).
далее алгоритмом бинарного поиска ищем занчения x при которых функция = 0.

```python
import sympy
x = sympy.Symbol('x')
f =  -12*x**4*sympy.sin(sympy.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30

def root(function):
    x_a = float(input('Введите значение левой границы поиска'))
    x_b = float(input('Введите значение правой границы поиска'))
    quession = int(input('введите "1" если на данном интервале функция убывающая и "2" - если возрастающая'))
    
    expression = False
    while (expression == False):
        delta = abs(abs(x_a) - abs(x_b)) / 2
        y_a = function.subs(x, (x_a + delta))
        if (round(y_a) > 0):
            if (quession == 1):
                x_a = x_a + delta
            else:
                x_b = x_a + delta
            # result_root = x_a
            expression = False
        elif (round(y_a) < 0):
            if (quession == 1):
                x_b = x_a + delta
            else:
                x_a = x_a + delta
            # result_root = x_a
            expression = False
        else:
            result_root = x_a + delta
            expression = True
    
    return result_root


# root(f)

```

ниже - функция поиска точек мининмума функции на указанных интервалах. Интервал задаем исходя из анализа графика.

```python
import sympy
x = sympy.Symbol('x')
f =  -12*x**4*sympy.sin(sympy.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30

def minimum(function):
    x_a = float(input('Введите значение левой границы поиска'))
    x_b = float(input('Введите значение правой границы поиска'))
   
    
    expression = False
    while (expression == False):
        delta = abs(abs(x_a) - abs(x_b)) / 2
        y_a = function.subs(x, (x_a + delta))
        y_test = function.subs(x, (x_a + delta + 0.001))
        if (y_a > y_test):
            x_a = x_a + delta
            expression = False
        elif (y_a < y_test):
            x_b = x_a + delta
            expression = False
        else:
            result_point = [x_a + delta, y_a]
            expression = True
    
    return result_point

# minimum(f)
```
