```python
import sympy
```

```python
x = sympy.Symbol('x')
```

```python
f =  5 * x * 6 * x + 4 * x + x * 6 + 8 + 9 + 4 * x * x + 5 * x
```

```python
print(f)
```

```python
print(sympy.solve(f, x))
```

```python
print([root.evalf() for root in sympy.solve(f, x)])
```
